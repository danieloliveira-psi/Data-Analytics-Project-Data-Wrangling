```python
tweet_errors={}
tweet_ids = list(df_twitter.tweet_id)

start = time.time()

with open('tweet_json.txt', 'w') as outfile:
    for tweet in tweet_ids:
        try:
            status = api.get_status(tweet, tweet_mode = 'extended')
            json.dump(status, outfile)
            outfile.write('\n')
            print('{}: {} (success!)'.format(tweet_ids.index(tweet), tweet))
        except Exception as e:
            tweet_errors[tweet] = e
            print('{}: {} (oops)'.format(tweet_ids.index(tweet), tweet))
            pass
                  
end = time.time()
duration = end-start
print('Can you believe this took {} seconds??. I mean, that\'s {} minutes long!'.format(duration, duration / 60))
print('Also, this is the number of tweet IDs that are not working: {}'.format(len(tweet_errors)))
```

## Analyze and Visualize Data

Let's do some brief and simple analysis and visualization of our data

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# First, some general information regarding our sample
df_twitter.describe()
```

```python
# Let's know the relative frequency of dog stages
df_twitter.groupby('dog_stage').tweet_id.count()/df_twitter.shape[0]*100
```

```python
# Now without the undefined cases (we'll create a temporary df to ease this visualization)
df_twitter_stages = df_twitter.query('dog_stage != "undefined"')

df_twitter_stages.groupby('dog_stage').tweet_id.count()/df_twitter_stages.shape[0]*100
```

```python
# I'll have a visual comparison between relative frequency of dog stages
df_twitter_stages.dog_stage.value_counts().plot(kind='bar')
plt.title('Number of Dog Stages')
plt.ylabel('Frequency')
plt.xlabel('Dog Stages');
```

```python
# What are their favorite counts
df_twitter_stages.groupby('dog_stage').favorite_count.mean()
```

```python
# And retweet counts
df_twitter_stages.groupby('dog_stage').retweet_count.mean()
```

```python
# What about their rating score
df_twitter_stages['rating_score'] = df_twitter_stages['rating_numerator'] / df_twitter_stages['rating_denominator']

df_twitter_stages.groupby('dog_stage').rating_score.mean()
```

```python
# How many dog breeds do we have? I'll do -1 because of unidentifiable cases
df_twitter.best_guess.nunique()-1
```

```python
# Let's see a scatter plot between favorite and retweet counts, I'm assuming they're related
df_twitter.plot(x='favorite_count', y='retweet_count', kind='scatter')
plt.title('Retweet Counts vs Favorite Counts')
plt.ylabel('Retweet Counts')
plt.xlabel('Favorite Counts');
```
